The purpose of this notebook is to review the impact of using various beta schedulers on the betas and alphas of the model

```python
from matplotlib import pyplot as plt

from diffusion import DiffusionSampler


def show_results(ds):
    print('alphas')
    alphas = ds.alphas.numpy()
    plt.plot(range(len(alphas)), alphas)
    plt.suptitle('Alphas')
    plt.show()
    alphas_cumproduct = ds.alphas_cumproduct.numpy()
    plt.plot(range(len(alphas_cumproduct)), alphas_cumproduct)
    plt.suptitle('Alphas cumulative')
    plt.show()
    print(f'alphas_cum {ds.alphas_cumproduct[-1].item():.4f}')
    print()
```

```python
beta_schedulers = [
    'cosine', 'linear', 'quadratic', 'sigmoid'
]

for beta_scheduler in beta_schedulers:
    print('=' * 60)
    ds = DiffusionSampler(
        beta_start=0.0001,
        beta_end=0.02,
        timesteps=300,
        beta_scheduler=beta_scheduler
    )
    show_results(ds)
    print()
```

```python
timesteps = range(100, 1001, 100)

for timestep in timesteps:
    print('=' * 60)
    ds = DiffusionSampler(
        beta_start=0.0001,
        beta_end=0.02,
        timesteps=timestep,
        beta_scheduler='cosine'
    )
    show_results(ds)
    print()
```

```python
ds = DiffusionSampler(
        beta_start=0.0001,
        beta_end=0.0065,
        timesteps=1000,
        beta_scheduler=beta_scheduler
    )
show_results(ds)
```
