The purpose of this notebook is to review the image data that gets passed to the model after data augmentations:

```python
import matplotlib.pyplot as plt

from dataset import SWImageDataModule

from utils import image_utils


def view_batch(batch, num_images=5):
    plt.figure(figsize=(15, 5))
    for i in range(num_images):
        img = batch[i]
        ax = plt.subplot(1, num_images, i + 1)
        image_utils.show_tensor_image(img.detach().cpu())
        ax.axis('off')    
    plt.show()


data_module = SWImageDataModule()
train_loader = data_module.train_dataloader()

num_views = 3

for _ in range(num_views):
    batch = next(iter(train_loader))
    view_batch(batch)
```

```python
from dataset import PoCDataModule

data_module = PoCDataModule()
train_loader = data_module.train_dataloader()

num_views = 3

for _ in range(num_views):
    batch, _ = next(iter(train_loader))
    view_batch(batch)
```
